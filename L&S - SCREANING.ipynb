1. Objetivo do Sistema
Identificar pares de ações com relacionamento estável no longo prazo (cointegradas), monitorar o spread entre elas e executar operações long/short quando o desvio estatístico excede bandas calculadas dinamicamente.
A estratégia explora o retorno à média de pares estáveis.

2. Coleta e Preparação de Dados
Download automático de preços ajustados via yfinance para mais de 100 ativos brasileiros.
Frequência: diária
Armazenamento: DataFrames do pandas, replicados para testes independentes dos pares.

3. Modelo Estatístico e Seleção de Pares
Para cada par analisado, o sistema calcula:

✓ 1. Cointegração — Teste ADF

O modelo estima o spread entre dois ativos via regressão linear.
Teste ADF no resíduo para validar estacionariedade.
Somente pares com significância elevada são aprovados.

✓ 2. Half-life (Meia-Vida)
Mede a velocidade de retorno do spread à média.
Utilizado como filtro adicional de qualidade do par.

✓ 3. Beta Rotation
Estima a estabilidade do beta ao longo do tempo.
Detecta pares cujo relacionamento varia (pair instability).

✓ 4. Faixas Estatísticas
Com base em:
média do spread
desvios padrão
último resíduo

São definidas bandas de ±1 DP e ±2 DP que atuam como gatilhos de entrada e saída.

4. Gestão de Risco e Dimensionamento das Ordens
O sistema possui um módulo de position sizing dinâmico, que considera:
Estatísticas do par (ADF, desvio padrão, estabilidade)
Faixas de preço dos ativos
Limites de tamanho (máximo 500 unidades por perna)

Quando ocorre sinal:
Compra de spread
Compra ativo 1
Vende ativo 2
Venda de spread
Vende ativo 1
Compra ativo 2

O fator multiplicador (1× a 5×) ajusta o tamanho da posição conforme a qualidade estatística do par.

5. Execução Real via MetaTrader 5
Quando um sinal válido é detectado:
Verifica se o símbolo está habilitado e visível no MT5.
Envia ordens a mercado para as duas pernas.
Registra o trade no arquivo Trades.csv.
Gera gráfico automático do spread no momento da execução.
Execução ocorre somente durante janela pré-determinada, evitando horários de baixa liquidez.

6. Backtesting Integrado
O sistema consulta automaticamente o arquivo Backtesting.csv e calcula:
Lucro acumulado histórico para o par
Taxa de acerto (% de trades positivos)
Isso permite priorizar pares com melhor performance histórica.

7. Monitoramento em Loop
O sistema roda em loop contínuo:
Avalia cada par
Atualiza estatísticas
Verifica gatilhos de entrada
Executa e registra trades
Encerra automaticamente ao atingir limites de horário ou número de operações

Este código implementa um módulo operacional responsável por monitorar, atualizar e encerrar posições abertas de uma estratégia de pairs trading. Ele atua como o “back-office inteligente” da estratégia: lê as operações em aberto, atualiza preços, calcula PnL, verifica regras de saída e envia ordens para zeragem no MetaTrader 5 de forma totalmente automatizada.


(1)O objetivo é:
✓ Monitorar continuamente os pares já negociados.
Carrega as informações do arquivo Trades.csv e identifica todos os ativos envolvidos.
✓ Atualizar automaticamente preços, valores de entrada e PnL.
Busca ticks em tempo real para calcular lucro/prejuízo de cada perna.
✓ Identificar trades que devem ser encerrados.

Utiliza dois gatilhos:
Half-life do spread: encerra quando o número de dias úteis supera a meia-vida estatística calculada no momento da entrada.
Janela de encerramento no pregão: entre 17:50 e 17:53, garantindo que nenhuma posição role para o próximo dia sem controle.
✓ Executar a zeragem das posições no MetaTrader 5.

O sistema automaticamente:
envia ordem de venda/compra para a primeira perna,
envia ordem correspondente para a segunda perna,
registra o fechamento no arquivo Resultados.csv
sincroniza o arquivo Trades.csv removendo posições finalizadas.

(2) Verificação e preparação dos ativos
Para cada ativo listado:
Confere visibilidade no MetaTrader 5
Força a habilitação do símbolo, se necessário
Coleta preços de bid/ask/last
Atualiza o preço de entrada diretamente no arquivo de trades, garantindo consistência histórica

(3) Cálculo de PnL em tempo real
Para cada trade aberto:
Obtém preços atuais das duas pernas
Calcula o lucro ou prejuízo de acordo com:
direção da operação (“Compra” ou “Venda”)
quantidade de cada ativo
diferença entre preço de entrada e preço atual
Resultado armazenado na coluna Profit.

(4) Regras de Encerramento
A posição é marcada para zeragem quando:

a) Half-life vencido
O sistema calcula quantos dias úteis se passaram desde o início do trade.
Se o número supera o half-life do par, a posição deve ser encerrada.

b) Janela obrigatória de encerramento diário
Entre 17:50 e 17:53, independentemente do half-life, o sistema fecha todas as operações pendentes.
Essa regra garante controle de risco intraday e evita carregamento acidental para o dia seguinte.

(5) Execução automática das ordens
Quando um trade deve ser zerado:
Envia a ordem contrária à posição original (buy/sell).
Envia simultaneamente a ordem contrária para o segundo ativo do par.
Registra o trade encerrado no dataset de resultados.
Remove a linha correspondente do dataset de trades abertos.

(6) Salvamento e Relatórios
Após processar todos os pares:
Atualiza Trades.csv (posições ainda abertas)
Atualiza Resultados.csv (posições encerradas)

Imprime:
lucro aberto total
lucro fechado total

Encerra a conexão com o MetaTrader 5.

```python
#SCREENING AÇÕES 2
import pandas as pd
import numpy as np
from datetime import datetime
import MetaTrader5 as mt5
from datetime import datetime
import chardet

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# VERIFICAR DISPONIBILIDADE DOS ATIVOS
def verificar_disponibilidade_ativos(symbol1):
    symbol1_info = mt5.symbol_info(symbol1)
    if symbol1_info is None:
        print(symbol1, 'Ativo 1 não encontrado')
    elif not symbol1_info.visible:
        print(symbol1, 'Ativo 1 não visível')
        if not mt5.symbol_select(symbol1, True):
            print(f'Symbol_select({symbol1}) falhou, tentando novamente...')
            time.sleep(2)  # Espera um pouco antes de tentar novamente
    # Se ambos os ativos estão visíveis, sai do loop
    if symbol1_info and symbol1_info.visible:
        print('Ativo visíveL')
        return symbol1    
    return symbol1

# DEFINIÇÃO DE PARAMETROS DOS ATIVOS
def ativos_parametros(symbol1):
    # PREÇOS
    lot1 = 100
    symbol1_ask = mt5.symbol_info_tick(symbol1).ask
    symbol1_bid = mt5.symbol_info_tick(symbol1).bid
    symbol1_last = mt5.symbol_info_tick(symbol1).last
    return symbol1_ask, symbol1_bid, symbol1_last, lot1

# COLETAR TRADES ABERTOS
def get_trades():
    trades = mt5.positions_get(symbol=symbol1)
    if trades is None:
        print("Nenhum trade: {}".format(mt5.last_error()))
        return None
    trade_data = []
    for trade in trades:
        trade_dict = {
            'ticket': trade.ticket,'time': trade.time,'type': trade.type,'volume': trade.volume,
            'price_open': trade.price_open,'sl': trade.sl,'tp': trade.tp,'price_current': trade.price_current,
            'swap': trade.swap,'profit': trade.profit,'symbol': trade.symbol,'comment': trade.comment,'external_id': trade.external_id}
        trade_data.append(trade_dict)
    trades = pd.DataFrame(trade_data)
    return trades

# O ENCERRAMENTO DAS OPERAÇÕES DEVE SER FEITO ENTRE AS 17:50 E 17:53
def passou_das_17_50():
    agora = datetime.now()
    hora_limite_inferior = agora.replace(hour=17, minute=50, second=0, microsecond=0)
    hora_limite_superior = agora.replace(hour=17, minute=53, second=59, microsecond=999999)
    return hora_limite_inferior <= agora <= hora_limite_superior

# Importar trades abertos e resultados das operacoes encerradas 
with open('Z:\\Trading\\Trades.csv', 'rb') as f: # Importar trades abertos
    result = chardet.detect(f.read())
    file_encoding = result['encoding']
trades_abertos = pd.read_csv('Z:\\Trading\\Trades.csv', encoding=file_encoding, on_bad_lines='skip')
with open('Z:\\Trading\\Resultados.csv', 'rb') as f: # Importar resultados fechados
    result = chardet.detect(f.read())
    file_encoding = result['encoding']
resultados = pd.read_csv('Z:\\Trading\\Resultados.csv', encoding=file_encoding, on_bad_lines='skip')
#trades_abertos = pd.read_csv('Z:\\Trading\\Trades.csv', on_bad_lines='skip')
#resultados = pd.read_csv('Z:\\Trading\\Resultados.csv', on_bad_lines='skip')

# Processar tickers
trades_abertos['Ativo1'] = trades_abertos['Ativo1'].str.replace('.SA', '', case=False)
trades_abertos['Ativo2'] = trades_abertos['Ativo2'].str.replace('.SA', '', case=False)
trades_abertos['Ativo1'] = trades_abertos['Ativo1'].str.upper()
trades_abertos['Ativo2'] = trades_abertos['Ativo2'].str.upper()

# Inicializar MetaTrader 5
if not mt5.initialize():
    print('Initialize() failed, error code = ', mt5.last_error())
    quit()

# Seleciona as colunas 'Ativo1' e 'Ativo2' e converte para listas
lista_ativos = trades_abertos[['Ativo1', 'Ativo2']].values.tolist()
lista_ativos = [item for sublist in lista_ativos for item in sublist]
lista_ativos = [item for item in lista_ativos if item != 'NAN']
display(lista_ativos)

# Atualizar preço de entrada dos ativos
for i in lista_ativos:
    symbol1 = i
    symbol1 = verificar_disponibilidade_ativos(symbol1)
    symbol1_ask, symbol1_bid, symbol1_last, lot1 = ativos_parametros(symbol1)
    trades = get_trades()
    trades = pd.DataFrame(trades)
    display(trades)
    try:
        price_open = float(trades.iloc[0, 4])
    except IndexError:
        pass
    trades_abertos.loc[trades_abertos['Ativo1'] == symbol1, 'PreçoEnt1'] = price_open
    trades_abertos.loc[trades_abertos['Ativo2'] == symbol1, 'PreçoEnt2'] = price_open

# Converter colunas para numéricos uma vez
cols_to_convert = ['PreçoExt1', 'PreçoEnt1', 'Qnt1', 'PreçoExt2', 'PreçoEnt2', 'Qnt2']
trades_abertos[cols_to_convert] = trades_abertos[cols_to_convert].apply(pd.to_numeric, errors='coerce')
#trades['HF'] = 0

# Processar trades
for indice, linha in trades_abertos.iloc[1:].iterrows():
    ativo1 = linha['Ativo1']
    ativo2 = linha['Ativo2']
    price1 = mt5.symbol_info_tick(ativo1).last
    price2 = mt5.symbol_info_tick(ativo2).last
    trades_abertos.at[indice, 'PreçoExt1'] = price1
    trades_abertos.at[indice, 'PreçoExt2'] = price2
    # Calcular PnL
    trade_type = linha['Trade']
    if trade_type == 'Compra':
        pnl_ativo1 = (price1 - linha['PreçoEnt1']) * linha['Qnt1']
        pnl_ativo2 = (linha['PreçoEnt2'] - price2) * linha['Qnt2']
        trades_abertos.at[indice, 'Profit'] = round(pnl_ativo1 + pnl_ativo2, 2)
        # Verificar se é hora de encerrar a posição
        data_trade = np.datetime64(linha['Data']).astype('datetime64[D]')
        half_life = int(linha['HF'])
        dias_uteis = np.busday_count(data_trade, np.datetime64('today'))
    
        if dias_uteis >= half_life:
            print('Existem trades para serem zerados hoje.')
            display(linha)
            if passou_das_17_50() or dias_uteis > half_life: # Os trades sao encerrados se for mais de 17:50 ou se a meia-vida venceu.
                # Enviar ordens para zerar posição
                lot1 = linha['Qnt1']
                point = mt5.symbol_info(linha['Ativo1']).point
                price = mt5.symbol_info_tick(linha['Ativo1']).last
                deviation = 5
                request1 = {'action': mt5.TRADE_ACTION_DEAL,
                    'symbol': linha['Ativo1'],'volume': float(lot1),
                    'type': mt5.ORDER_TYPE_SELL,
                    'price': price,
                    'deviation': deviation,'magic': 234000,'comment': 'python script open','type_time': mt5.ORDER_TIME_GTC,'type_filling': mt5.ORDER_FILLING_RETURN,}
                
                lot2 = linha['Qnt2']
                point = mt5.symbol_info(linha['Ativo2']).point
                price = mt5.symbol_info_tick(linha['Ativo2']).last
                deviation = 5
                request2 = {'action': mt5.TRADE_ACTION_DEAL,
                    'symbol': linha['Ativo2'],'volume': float(lot2),
                    'type': mt5.ORDER_TYPE_BUY,
                    'price': price,
                    'deviation': deviation,'magic': 234000,'comment': 'python script open','type_time': mt5.ORDER_TIME_GTC,'type_filling': mt5.ORDER_FILLING_RETURN,}
                result1 = mt5.order_send(request1)
                #print(result1)
                result2 = mt5.order_send(request2)
                #print(result2)
                # Adicionar trade a resultados
                resultados = pd.concat([resultados, pd.DataFrame([linha])], ignore_index=True)
                trades_abertos.drop(indice, inplace=True)

    elif trade_type == 'Venda': # Venda
        pnl_ativo1 = (linha['PreçoEnt1'] - price1) * linha['Qnt1']
        pnl_ativo2 = (price2 - linha['PreçoEnt2']) * linha['Qnt2']
        trades_abertos.at[indice, 'Profit'] = round(pnl_ativo1 + pnl_ativo2, 2)
        # Verificar se é hora de encerrar a posição
        data_trade = np.datetime64(linha['Data']).astype('datetime64[D]')
        half_life = int(linha['HF'])
        dias_uteis = np.busday_count(data_trade, np.datetime64('today'))
    
        if dias_uteis >= half_life:
            print('Existem trades para serem zerados hoje.')
            display(linha)
            if passou_das_17_50() or dias_uteis > half_life:
                # Enviar ordens para zerar posição
                lot1 = linha['Qnt1']
                point = mt5.symbol_info(linha['Ativo1']).point
                price = mt5.symbol_info_tick(linha['Ativo1']).last
                deviation = 5
                request1 = {'action': mt5.TRADE_ACTION_DEAL,
                    'symbol': linha['Ativo1'],'volume': float(lot1),
                    'type': mt5.ORDER_TYPE_BUY,
                    'price': price,
                    'deviation': deviation,'magic': 234000,'comment': 'python script open','type_time': mt5.ORDER_TIME_GTC,'type_filling': mt5.ORDER_FILLING_RETURN,}
                
                lot2 = linha['Qnt2']
                point = mt5.symbol_info(linha['Ativo2']).point
                price = mt5.symbol_info_tick(linha['Ativo2']).last
                deviation = 5
                request2 = {'action': mt5.TRADE_ACTION_DEAL,
                    'symbol': linha['Ativo2'],'volume': float(lot2),
                    'type': mt5.ORDER_TYPE_SELL,
                    'price': price,
                    'deviation': deviation,'magic': 234000,'comment': 'python script open','type_time': mt5.ORDER_TIME_GTC,'type_filling': mt5.ORDER_FILLING_RETURN,}
                
                result1 = mt5.order_send(request1)
                #print(result1)
                result2 = mt5.order_send(request2)
                #print(result2)
                # Adicionar trade a resultados
                resultados = pd.concat([resultados, pd.DataFrame([linha])], ignore_index=True)
                trades_abertos.drop(indice, inplace=True)

# Remover duplicatas em resultados
#resultados.drop_duplicates(subset=['Data', 'Trade', 'Ativo1', 'Ativo2', 'PreçoEnt1','PreçoEnt2','ADF'], keep='first', inplace=True)

resultados = resultados.iloc[:, :15]

# Salvar resultados
resultados.to_csv('Z:\\Trading\\Resultados.csv', index=False)
trades_abertos.to_csv('Z:\\Trading\\Trades.csv', index=False)

# Finalizar MT5
mt5.shutdown()
display(trades_abertos)
display(resultados)
print(' ')
print('Lucro em aberto: ',trades_abertos['Profit'].sum())
print(' ')
print('Lucro fechado: ',resultados['Profit'].sum())
print(' ')
```
